```python
def change_device(df):
    """
    change the devices to a single word to represent it, group the different tweet sources
    """
    def change(device_line):
        if 'Twitter Web Client' in device_line:
            return 'twitter_web'
        elif 'Twitter for BlackBerry' in device_line:
            return 'twitter_bb'
        elif 'Twitter for iPad' in device_line:
            return 'twitter_ipad'
        elif 'TweetDeck' in device_line:
            return 'twitter_deck'
        elif 'instagram' in device_line:
            return 'instagram'
        elif 'periscope' in device_line:
            return 'periscope'
        elif 'Facebook' in device_line:
            return 'facebook'
        else:
            return device_line
    df["device"] = df["device"].apply(change)
    return df
```

### feature extraction

i plan to extract features numerical faetures about the text, the number of offensive words has been used in them.  
the offensive words comes from: https://github.com/LDNOOBW/List-of-Dirty-Naughty-Obscene-and-Otherwise-Bad-Words/blob/master/en, https://gist.github.com/jamiew/1112488

```python
def get_offensive1():
    """
    get the words from the file, only single words. filter combinations
    """
    words = []
    with open("data/offensive1.txt",'r') as f:
        words = [x.strip("\n") for x in f if " " not in x]
    return words
```

```python
def get_offensive2():
    """
    get the words from the file, only single words. filter combinations
    """
    words = []
    with open("data/offensive2.txt",'r') as f:
        for line in f:
            line = line.strip("\n")
            parts = line.split(":")
            if len(line)!=0:
                word = parts[0].strip('"').strip(" ")
                if " " not in word:
                    words.append(word)
    return words
```

combine the offensive words

```python
offensive_words = set(get_offensive1()+get_offensive2())
```

```python
def extract_features(x):
    """
    extract the average/min/max for the word and sentences lenghts
    """
    sentences = sent_tokenize(x)
    words = word_tokenize(x)
    
    senteces_sizes =[len(word_tokenize(sentence)) for sentence in sentences]
    words_sizes = [len(word) for word in words]
    num_of_sentences = len(sentences)
    num_of_words = len(words)
    avg_sentences_length = np.mean(senteces_sizes)
    avg_words_length = np.mean(words_sizes)
    if len(words_sizes)>0:
        min_word_length = np.min(words_sizes)
        max_word_length = np.max(words_sizes)
    else:
        min_word_length = 0
        max_word_length = 0
    if len(senteces_sizes)>0:
        min_sentence_length = np.min(senteces_sizes)
        max_sentence_length = np.max(senteces_sizes)
    else:
        min_sentence_length = 0
        max_sentence_length = 0
    num_off_words = sum([1 for word in words if word in offensive_words])
    num_of_cap = sum([1 for word in words if word.isupper()])
    return np.asarray([num_of_sentences,min_sentence_length,avg_sentences_length,max_sentence_length,num_of_words,min_word_length,avg_words_length,max_word_length,num_off_words,num_of_cap])
```

```python
def plot_algorithms_results(name,clf,height=None):
    labels = tuple(map(lambda x: str(x),clf.cv_results_["params"]))
    acc = tuple(clf.cv_results_["mean_test_score"])
    std = tuple(clf.cv_results_["std_test_score"])

    N = len(labels)

    ind = np.arange(N)  # the x locations for the groups
    width = 0.7       # the width of the bars

    plt.rcdefaults()
    fig, ax = plt.subplots()
    if height is not None:
        fig.set_figheight(height)
    
    rects = ax.barh(ind, acc, xerr=std, align='center',
        color='blue', ecolor='black')
    ax.set_yticks(ind)
    ax.set_yticklabels(labels)
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel('Accucary')
    ax.set_title(name)
    
    for i, v in enumerate(acc):
        ax.text(rects[0].get_width()*1.08, i +.1, str(v),color="black")
        
    plt.show()
```

```python
def preprocess(x):
    regex = re.compile('[^a-zA-Z ]')
    return regex.sub('', str(x)).lower()
#     return str(x).replace("\n","").replace("\r","").replace("."," ").replace(","," ").lower()
```

```python
def topic_model(documents,no_features,no_topics,tf=None,tf_vectorizer=None):
    if tf is None and tf_vectorizer is None:
        tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2, max_features=no_features, stop_words='english')
        tf = tf_vectorizer.fit_transform(documents)
    tf_feature_names = tf_vectorizer.get_feature_names()
    lda = LatentDirichletAllocation(n_components=no_topics, max_iter=5, learning_method='online', learning_offset=50.,random_state=0,n_jobs=-1).fit(tf)
    return lda,tf_feature_names,tf,tf_vectorizer
```

```python
def get_topics(model, feature_names, no_top_words):
    topics = {}
    for topic_idx, topic in enumerate(model.components_):
        topics[topic_idx] = [feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]
    return topics
```

```python
def make_models(min_topics,max_topics,no_features,train_data,test_data):
    tf = None
    tf_vectorizer = None
    perplexities = {}
    topics = {}
    models = {}
    for i in range(min_topics,max_topics):
        print "preparing model with %s topics"%i
        model1,feature_names1,tf,tf_vectorizer = topic_model(train_data,no_features,i,tf,tf_vectorizer)
        perplexity = model1.perplexity(tf_vectorizer.transform(test_data))
        perplexities[i]=perplexity
        topics[i] = get_topics(model1,feature_names1,15)
        models[i] = model1
    return perplexities,topics,models,feature_names1
```

```python
def make_kmeans_models(from_n,to_n,data):
    models = {}
    for n in range(from_n,to_n):
        model = KMeans(n_clusters=n,n_jobs=-1).fit(data)
        labels = model.labels_
        sil_score = silhouette_score(data, labels, metric='euclidean')
        cal_score = calinski_harabaz_score(data, labels)
        models[n] = model,sil_score,cal_score
    return models
```

```python
def make_kmeans_models2(from_n,to_n,data):
    models = {}
    for n in range(from_n,to_n):
        model = KMeans(n_clusters=n,n_jobs=-1).fit(data)
        labels = model.labels_
#         sil_score = silhouette_score(data, labels, metric='euclidean')
        cal_score = calinski_harabaz_score(data, labels)
        models[n] = model,cal_score
    return models
```

```python
def display_clusters(models,data,name):
    plt.figure(figsize=(15, 50))
    plt.subplots_adjust(left=.02, right=.98, bottom=.001, top=.96, wspace=.05,
                    hspace=.01)
    pca = PCA(n_components=2)
    pca.fit(data)
    data = pca.transform(data)
    for i,key in enumerate(models.keys()):
        labels = models[key][0].labels_.astype(np.int)
        plt.subplot(len(models), 3, i+1)
        if i == 0:
            plt.title(name, size=18)
        colors = np.array(list(islice(cycle(['#377eb8', '#ff7f00', '#4daf4a',
                                             '#f781bf', '#a65628', '#984ea3',
                                             '#999999', '#e41a1c', '#dede00']),
                                      int(max(labels) + 1))))
        plt.scatter(data[:, 0], data[:, 1], s=10, color=colors[labels])

        plt.xticks(())
        plt.yticks(())
        plt.text(.99, .01, key,
                 transform=plt.gca().transAxes, size=15,
                 horizontalalignment='right')
    plt.show()
```
